```python
import pandas as pd
```

```python
elementary = pd.read_csv('2세종시 초등학교현황.csv',encoding='cp949')
elementary.head()
```

```python
# 주소만 가져오자.
adds = elementary["주소"]
adds.head()
```

```python
# list로 위도경도 싹 받아주자.
latitude = []
longitude = []
locator = Nominatim(user_agent="myGeocoder")

for add in adds:
  location = locator.geocode(add)
  latitude.append(location.latitude)
  longitude.append(location.longitude)
```

```python
# latitude,longitude가 list 형태로 받아졌다
type(latitude)
```

```python
# 이제 위에 elementary에 추가한다.
elementary["latitude"] = latitude
elementary["longitude"] = longitude
```

```python
elementary.head()
```

```python
# 이제 지도 그리자
import folium
import numpy as np
```

```python
# 세종시 조치원 청사 위도 경도
latitude  = 36.596650401760954
longitude = 127.29238947025198
loc_tuple = (latitude, longitude)
loc_tuple

map = folium.Map(
    location=loc_tuple,
    tiles='cartodbpositron',
    zoom_start=11,
)

# CircleMarker 로 초등학교 위치 표시.
elementary.apply(lambda row:folium.CircleMarker(location=[row["latitude"], row["longitude"]]).add_to(map), axis=1)
map
```
